# Tutorial de Graph Tools

Instalación en Ubuntu, tomado de la [documentación oficial](https://git.skewed.de/count0/graph-tool/wikis/installation-instructions#debian-ubuntu):

- Agregar en la lista de repositorios `/etc/apt/sources.list` las siguientes líneas (es necesario ser superusuario para hacer esto):
```
deb http://downloads.skewed.de/apt/DISTRIBUTION DISTRIBUTION universe
deb-src http://downloads.skewed.de/apt/DISTRIBUTION DISTRIBUTION universe
```

    Donde DISTRIBUTION es cualquiera de las siguientes: `xenial, yakkety, zesty, artful, bionic, cosmic`

- Agregar la clave pública para la verificación de los paquetes:
```
apt-key adv --keyserver pgp.skewed.de --recv-key 612DEFB798507F25
```

- Actualizar la lista con ```sudo apt update```
- Instalar graph-tools con ```sudo apt install python-graph-tool```

## Creación de redes sencillas

````python
from graph_tool import Graph
````

````python
# Grafo no dirigido
g = Graph(directed=False)
````

````python
# Agregado de nodos 
v0 = g.add_vertex()
v1 = g.add_vertex()
v2 = g.add_vertex()

# Agregado de enlaces
e0 = g.add_edge(v0,v1)
e1 = g.add_edge(v0,v2)

# Agregar varios nodos, por ejemplo 3
vlist = g.add_vertex(3)

# Remover nodos, por ejemplo el 4
g.remove_vertex(4)

# Ver los nodos que hay en la red, el índice los ordena en forma correlativa
for v in g.vertices():
    print(v)
    
````

````python
# Visualización 
from graph_tool.draw import graph_draw

graph_draw(g)
````

### Propiedades de la red

````python
# Número de nodos
print('#Nodos: {}'.format(g.num_vertices()))

# Número de enlaces
print('#Enlaces: {}'.format(g.num_edges()))
````

### Propiedades de los nodos

````python
# Grado del nodo (para redes no dirigidas el grado de un nodo es igual a su out_degree)
print('#Grado del nodo 1: {}'.format(g.vertex(0).out_degree()))

# Para agregar alguna propiedad a cada nodo, tenemos que declarar la variable
v_age = g.new_vertex_property("int")

# Esta variable queda como un diccionario en el que seteamos 1 a 1 las propiedades
v_age[0] = 24
v_age[1] = 28

# O bien hacemos un for
for v in g.vertices():
    v_age[v] = 43

# Se lo asignamos al grafo
g.vertex_properties["age"] = v_age

# Para obtener las propiedades de cada nodo
print(g.vertex_properties["age"][0])
````

### Propiedades de los enlaces

````python
# Fuente y objetivo de un determinado enlace

print('Fuente: {}'.format(e0.source()))
print('Fuente: {}'.format(e0.target()))
````

````python
import numpy as np
weight = g.new_edge_property("float")

for e in g.edges():
    weight[e] = np.random.random()

g.edge_properties["weight"] = weight
````

````python
g.save('prueba.gml')
````
